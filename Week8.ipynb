```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('/content/drive/MyDrive/Image/j.png')
plt.imshow(img)
```

```python
# Task 3
kernel = np.ones((5,5), np.uint8)
dilation = cv2.dilate(img, kernel, iterations = 1)

plt.subplot(121),plt.imshow(img),plt.title('Original image')
plt.subplot(122),plt.imshow(dilation),plt.title('Dilated image')
```

```python
#Task 4
kernel = np.ones((5,5), np.uint8)
erosion = cv2.morphologyEx(img, cv2.MORPH_ERODE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Original image')
plt.subplot(122), plt.imshow(erosion), plt.title('Image after erosion')
```

```python
#Task 5
kernel = np.ones((5,5), np.uint8)
dilation = cv2.morphologyEx(img, cv2.MORPH_DILATE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Original image')
plt.subplot(122), plt.imshow(dilation), plt.title('Image after dilation')
```

```python
#Task 6
kernel = np.ones((5,5), np.uint8)
erosion = cv2.morphologyEx(img, cv2.MORPH_ERODE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Original image')
plt.subplot(122), plt.imshow(erosion), plt.title('Image after Erosion')
```

```python
# Task 7 using library
kernel = np.ones((7,7), np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Original image')
plt.subplot(122), plt.imshow(opening), plt.title('Image after opening')
```

```python
#Task 7 without using library

img = cv2.imread('/content/drive/MyDrive/Image/j.png')

element = cv2.getStructuringElement(cv2.MORPH_RECT,(7,7))
erosion_dst = cv2.erode(img, element)
opening_dst = cv2.dilate(erosion_dst, element)

plt.subplot(121), plt.imshow(img), plt.title('Original image')
plt.subplot(122), plt.imshow(opening_dst), plt.title('Image after Opening')
```

```python
#Task 8 using library

kernel = np.ones((7,7), np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Original image')
plt.subplot(122), plt.imshow(closing), plt.title('Image after closing')
```

```python
#Task 8 without using library

img = cv2.imread('/content/drive/MyDrive/Image/j.png')

element = cv2.getStructuringElement(cv2.MORPH_RECT,(7,7))
dilate_dst = cv2.dilate(img, element)
closing_dst = cv2.erode(dilate_dst, element)

plt.subplot(121), plt.imshow(img), plt.title('Original image')
plt.subplot(122), plt.imshow(closing_dst), plt.title('Image after closing')
```
